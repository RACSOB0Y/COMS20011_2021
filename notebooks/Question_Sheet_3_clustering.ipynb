<h3> Question 2: K means M-step </h3>

For the following data-points, and cluster assignments, $z$, perform one M-step to find the corresponding cluster-centers.

```  
 x_0    x_1   z
-2.1   -3.2   1
-3.4   -1.2   0
-2.6   -2.7   1
 3.2    2.1   0
 1.2    3.6   1
 0.6    0.0   0
```



````python
def mean(a):
    s = 0
    for d in a:
        s+=d
    return s/len(a)


def m_step(dat):
    d_0 = []
    d_1 = []
    for d in dat:
        if d[2] == 0:
            d_0.append(d[0:2])
        else:
            d_1.append(d[0:2])
    d_0 = np.array(d_0)
    d_1 = np.array(d_1)
    return ([mean(d_0[:,0]), mean(d_0[:,1])], [mean(d_1[:,0]), mean(d_1[:,1])])


d = np.array([[-2.1, -3.2, 1],
    [-3.4, -1.2, 0],
    [-2.6, -2.7, 1],
    [3.2, 2.1, 0],
    [1.2, 3.6, 1],
    [0.6, 0.0, 0]])
(m_0, m_1) = m_step(d)
Y = d[:,2]
d_0 = []
d_1 = []
for p in d:
    if p[2] == 0:
        d_0.append(p[0:2])
    else:
        d_1.append(p[0:2])
d_0 = np.array(d_0)
d_1 = np.array(d_1)

fig, ax = plt.subplots()
ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)
ax.scatter(m_0[0], m_0[1])
ax.scatter(m_1[0], m_1[1])
ax.scatter(d_0[:,0], d_0[:,1])
ax.scatter(d_1[:,0], d_1[:,1])
````

<h3> Question 3: K means E-step </h3>

Given cluster centers in the previous question, update the cluster assignments.

````python
for i in range(len(d)):
    if math.sqrt((m_0[0]-d[i][0])**2 + (m_0[1]-d[i][1])**2) < math.sqrt((m_1[0]-d[i][0])**2 + (m_1[1]-d[i][1])**2):
        d[i][2] = 0
    else:
        d[i][2] = 1
        
d_0 = []
d_1 = []
for p in d:
    if p[2] == 0:
        d_0.append(p[0:2])
    else:
        d_1.append(p[0:2])
d_0 = np.array(d_0)
d_1 = np.array(d_1)
print(d)
fig, ax = plt.subplots()
ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)
ax.scatter(m_0[0], m_0[1])
ax.scatter(m_1[0], m_1[1])
ax.scatter(d_0[:,0], d_0[:,1])
ax.scatter(d_1[:,0], d_1[:,1])
````

<h3> Question 4: Bayesian classification </h3>

Given training data 
```  
 x_0   y
-2.1   0
-3.4   0
-2.6   0
 3.2   1
 1.2   1
 0.6   1
```

Fit Gaussian distributions to each cluster using maximum-likelihood, then compute the corresponding posterior over the class-label for $x_0=-0.3$ (specifically $\P{y=1| x_0=-0.3}$).

(Note the answer to this question is in the previous week's answers as question 6).

````python
x = [-2.1, -3.4, -2.6, 3.2, 1.2, 0.6]
x_0 = x[0:3]
x_1 = x[3:]
x_data = np.arange(-4, 4, 0.001)

y0_data = stats.norm.pdf(x_data, mean(x_0), statistics.variance(x_0))
y1_data = stats.norm.pdf(x_data, mean(x_1), statistics.variance(x_1))

p0 = stats.norm.pdf(-0.3, mean(x_0), statistics.variance(x_0))
p1 = stats.norm.pdf(-0.3, mean(x_1), statistics.variance(x_1))
print(f"0 : m = {mean(x_0)} v = {statistics.variance(x_0)}\n\tp(-0.3|0) = {p0}")
print(f"1 : m = {mean(x_1)} v = {statistics.variance(x_1)}\n\tp(-0.3|1) = {p1}")

print(f"ML 0 = {p0/(p0+p1)}\nML 1 = {p1/(p0+p1)}")

fig, ax = plt.subplots()
ax.scatter(-0.3,0, c='g')
ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)
ax.scatter(x_0, t.zeros(3), c='r')
ax.plot(x_data,y0_data, c='r')
ax.scatter(x_1, t.zeros(3), c='b')
ax.plot(x_data,y1_data, c='b')
````
