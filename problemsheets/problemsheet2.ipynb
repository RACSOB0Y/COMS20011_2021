### 1 - Using sin(2πnx), demonstrate the concept of superposition as follows (in Matlab or Python):
(a) first plot three sine functions over the range ±3 in steps of 0.1 using n={1/4,1,2}. Note, plots should
appear in the same graph to give a better sense of what is happening.

(b) Now plot in a different colour the sum of all the sines above.

(c) Add more sine functions over the same range and repeat step (b).

```python
x = np.linspace(-3,3, 1000)
y0 = np.sin(2 * np.pi * 0.25 * x)
y1 = np.sin(2 * np.pi * 1 * x)
y2 = np.sin(2 * np.pi * 2 * x)
y3 = np.sin(2 * np.pi * 0.5 * x)
fig0, ax0 = plt.subplots()
plt.xlabel("θ")
plt.ylabel("sin(2πnθ)")
ax0.plot(x, y0, c='r', label='n = 0.25', alpha=0.3)
ax0.plot(x, y1, c='g', label='n = 1', alpha=0.3)
ax0.plot(x, y2, c='b', label='n = 2', alpha=0.3)
ax0.plot(x, y3, c='y', label='n = 0.5', alpha=0.3)
ax0.plot(x, y0 + y1 + y2 + y3, c='black', label='sum')
plt.legend()
plt.show()
```
